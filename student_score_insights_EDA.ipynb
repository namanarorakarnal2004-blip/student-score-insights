## 1️⃣ Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

sns.set_theme(style='whitegrid', palette='muted')
plt.rcParams['figure.figsize'] = (10, 5)
print('✅ Libraries loaded!')
```

## 2️⃣ Load Dataset

```python
df = pd.read_excel('Cleaned_Complete_Techno_Sales_Data.xlsx')
print(f'Shape: {df.shape}')
df.head()
```

## 3️⃣ Basic Info & Missing Values

```python
print('--- Data Types ---')
print(df.dtypes)
print()
print('--- Missing Values ---')
print(df.isnull().sum())
```

```python
# Fix: Drop rows with null Category
df = df.dropna(subset=['Category'])

# Fix: Standardize 'MotherBoard' vs 'Motherboard'
df['Category'] = df['Category'].replace('MotherBoard', 'Motherboard')

print(f'Clean dataset shape: {df.shape}')
```

## 4️⃣ Descriptive Statistics

```python
df[['Cost', 'Sales', 'Quantity', 'Total_Cost', 'Total_Sales']].describe().round(2)
```

## 5️⃣ Sales by Category

```python
cat_sales = df.groupby('Category')['Total_Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 5))
sns.barplot(x=cat_sales.index, y=cat_sales.values, palette='viridis')
plt.title('Total Sales by Category', fontsize=14)
plt.xlabel('Category')
plt.ylabel('Total Sales (₹)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 6️⃣ Sales by Brand

```python
brand_sales = df.groupby('Brand')['Total_Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 5))
sns.barplot(x=brand_sales.index, y=brand_sales.values, palette='rocket')
plt.title('Total Sales by Brand', fontsize=14)
plt.xlabel('Brand')
plt.ylabel('Total Sales (₹)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 7️⃣ Order Status Distribution

```python
status_counts = df['Status'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(status_counts.values, labels=status_counts.index, autopct='%1.1f%%',
        colors=sns.color_palette('pastel'), startangle=140)
plt.title('Order Status Distribution', fontsize=14)
plt.tight_layout()
plt.show()
```

## 8️⃣ Top 10 States by Revenue

```python
state_sales = df.groupby('State_Code')['Total_Sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 5))
sns.barplot(x=state_sales.index, y=state_sales.values, palette='coolwarm')
plt.title('Top 10 States by Total Sales', fontsize=14)
plt.xlabel('State Code')
plt.ylabel('Total Sales (₹)')
plt.tight_layout()
plt.show()
```

## 9️⃣ Profit Margin Analysis

```python
df['Profit'] = df['Total_Sales'] - df['Total_Cost']
df['Profit_Margin_%'] = ((df['Profit'] / df['Total_Sales']) * 100).round(2)

category_profit = df.groupby('Category')['Profit'].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 5))
sns.barplot(x=category_profit.index, y=category_profit.values, palette='Set2')
plt.title('Total Profit by Category', fontsize=14)
plt.xlabel('Category')
plt.ylabel('Profit (₹)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print(f'Overall Avg Profit Margin: {df["Profit_Margin_%"].mean():.2f}%')
```

## 🔟 Correlation Heatmap

```python
plt.figure(figsize=(8, 6))
corr = df[['Cost', 'Sales', 'Quantity', 'Total_Cost', 'Total_Sales', 'Profit']].corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', square=True)
plt.title('Correlation Heatmap', fontsize=14)
plt.tight_layout()
plt.show()
```

## 1️⃣1️⃣ Top Supervisors by Revenue

```python
sup_sales = df.groupby('Assigned Supervisor')['Total_Sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 5))
sns.barplot(x=sup_sales.values, y=sup_sales.index, palette='magma')
plt.title('Top 10 Supervisors by Total Sales', fontsize=14)
plt.xlabel('Total Sales (₹)')
plt.ylabel('Supervisor')
plt.tight_layout()
plt.show()
```
